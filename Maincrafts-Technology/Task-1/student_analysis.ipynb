# **Student Performance Analysis**
​This project involves exploring the Student Performance dataset to analyze factors influencing final grades (G3).
 We follow a data science workflow: Load → Clean → Analyze → Visualize → Conclude.

# **Load Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#use pandas to read dataset
df= pd.read_csv("student-mat.csv",sep=";")
df.head()
```

# **Data Cleaning**
We check for missing values and duplicates to ensure the data is accurate before analysis.

## **Explore & Clean Data**

```python
#chech missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#inspect dataset shape & dtypes
print("\nDataTypes :\n",df.dtypes)
```

## **Analysis Questions**
We are calculating the average grade, identifying top performers, and checking the correlation between study habits and results.

### **Average final grade (G3)**

```python
avg_grade = df['G3'].mean()
print("Average Final Grade:", avg_grade)
```


### **How many students scored above 15?**

```python
high_scorers = df[df['G3'] > 15].shape[0]
print("Number of Students with Score Above 15:", high_scorers)
```

### **Is study time correlated with performance?**

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between Study Time and Performance:", correlation)
```


### **Which gender performs better on average?**

```python
gender_performance = df.groupby('sex')['G3'].mean()
print("Average Performance by Gender:\n", gender_performance)
```

# **Visualization**
Graphical representations of the grade distribution and relationships between variables

### **Histogram of grades**


```python
plt.hist(df['G3'],bins=10,color='skyblue',edgecolor='black')
plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Frequency')
```

### **Scatterplot: study time vs grades**

```python
sns.scatterplot(x='studytime',y='G3',data=df)
plt.title('Study Time vs Final Grades')
```

### **Bar chart: male vs female average score**

```python
sns.barplot(x='sex',y='G3',data=df)
plt.title('Average Grade Male vs Female')
```

# **Conclusion**
Based on the analysis, we observed the distribution of student grades and identified how study time correlates with academic success. This workflow demonstrates the fundamental steps of a data science project.
